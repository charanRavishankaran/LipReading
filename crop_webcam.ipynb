```python
import dlib
video = cv2.VideoCapture(0)
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
pad1 = 12
pad2 = 24
a = []
#label=["Begin", "Choose", "Connection", "Navigation", "Next", "Previous", "Start", "Stop", "Hello", "Web"]
#label = ["Web","Web","Web"]
label = ["Begin", "Choose","Next","Start","Stop","Web"]
test=[]
temp=""
while True:
    check, frame = video.read()
    #print (check)
    #print (frame)
    img = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    #print("Size {}".format(img.shape))
    dets = detector(img, 1)
    for k, d in enumerate(dets):
        shape = predictor(img, d)
        xmouthpoints = [shape.part(x).x for x in range(48,67)]
        ymouthpoints = [shape.part(x).y for x in range(48,67)]
        maxx = max(xmouthpoints)
        minx = min(xmouthpoints)
        maxy = max(ymouthpoints)
        miny = min(ymouthpoints)
        avgx = round((maxx+minx)/2)
        avgy = round((maxy+miny)/2)
        crop_image = img[avgy-pad1:avgy+pad1,avgx-pad2:avgx+pad2]
        cv2.imwrite('lip.jpg',crop_image)
        frames_path = glob.glob ("C:/Users/Shyamprasad/Desktop/CV/Test/Code/webcam_test/lip.jpg")
        for x in frames_path:
            read = cv2.imread(x)
            a.append(read)
            if ((len(a))==10):
                print("Reading lip movements...")
                print("###################################################")
                testnp = np.array(a)
                x = np.expand_dims(testnp, axis=0)
                predictions = history.predict(x)
                temp = label[np.argmax(predictions)]
                print("Work spoken - ",temp)
                a=[]
        cv2.putText(frame,temp, (35, 50), cv2.FONT_HERSHEY_SIMPLEX,1.25, (0, 0, 255), 5)
    cv2.imshow("capturing",frame)
    cv2.imshow("Cropping lips",crop_image)    
    key = cv2.waitKey(1)
    if key==ord('q'):
        break  
    
        #cv2.waitKey(0)

video.release()
cv2.destroyAllWindows()
```
